### Transform FTSE industry into categories

```python
mapping = {'basic materials':0, 'consumer discretionary':1, 'consumer staples':2,
           'energy':3, 'financials':4, 'health care':5, 'industrials':6, 
           'real estate':7, 'technology':8, 'telecommunications':9, 'utilities':10}
```

```python
data['Industry cat']=data['FTSE industry'].replace(mapping)
```

```python
mapping_colors = {0:"rgb(255, 0, 0)", 1:"rgb(255, 83, 0) ", 2:"rgb(255, 165, 0)", 3:"rgb(255, 210, 0)", 4:"rgb(255, 255, 0)", 5:"rgb(128, 192, 0)", 6:"rgb(0, 128, 0)", 7:"rgb(0, 64, 128)", 8:"rgb(0, 0, 255)",
                  9:"rgb(38, 0, 193)",10:"rgb(75, 0, 130)"}
```

```python
data['data_colors'] = data['Industry cat'].replace(mapping_colors)
```

```python
#data['rank_2019'] = pd.qcut(data['2019'],5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
data['rank_2019'] = pd.qcut(data['2019'],5, labels=[15, 14, 13, 12, 11])
data['rank_2020'] = pd.qcut(data['2020'],5, labels=[20, 19, 18, 17, 16])
data['rank_2021'] = pd.qcut(data['2021'],5, labels=[25, 24, 23, 22, 21])
```
