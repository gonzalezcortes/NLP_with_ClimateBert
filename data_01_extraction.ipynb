```python

count=1
for subdir in os.listdir(directory):
  d = os.path.join(directory, subdir)
  for filename in os.listdir(d):
    file_name=os.path.join(d,filename)

    print(count)
    print('\n')
    print(filename+" "+subdir)
    print('\n')
    count=count+1
    
    tfname=filename+".json"

    outputfolder= '/content/drive/My Drive/Research Papers/Climate_2022/Output Files2/'
    tfdir=os.path.join(outputfolder,tfname)
    
    if os.path.isfile(tfdir) == True: continue
    
    company=""
    year=""
    typeof=""
    words=filename.split('_')
    company=words[0]
    year=words[-1]
    if(len(words)!=2): 
      typeof=words[1]

    doc_name=company+'_'+typeof+'_'+year+".json"

    check='/content/drive/My Drive/Research Papers/Climate_2022/Output Files2/' + doc_name 

    if(os.path.isfile(check)): continue
    
    with pikepdf.Pdf.open(file_name, password='') as pdf:
      pdf.save('output.pdf')
    doc=PyPDF2.PdfFileReader('/content/output.pdf')
    pages=doc.getNumPages()

    text=""
    for page in range(pages):
      pg=doc.getPage(page)
      text1=pg.extractText()
      text=text+text1

    sent=nltk.sent_tokenize(text)
    
    for t in sent:
      if len(t.split())<5:
        sent.remove(t)

    text=text.replace('\n',' ')
    sent2=nltk.sent_tokenize(text)
    doc=[]
    for i in sent2:
      dic={"Company":company, "Text":i}
      doc.append(dic)

    with open(doc_name,'w') as f:
        json.dump(doc, f, indent=3)
        #json.dump(json_string,f,indent=3)

    doc_name=company+'_'+typeof+'_'+year+".json" 
    fn=os.path.join("/content",doc_name)
    !cp $fn '/content/drive/My Drive/Research Papers/Climate_2022/Output Files2/'
```
